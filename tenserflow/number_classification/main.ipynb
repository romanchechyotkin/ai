```python
import tensorflow as tf
import numpy as np 

from tensorflow import keras
from tensorflow.keras.preprocessing import image
```

```python
print(tf.__version__)
```

```python
(train_images, train_labels), (test_images, test_labels) = keras.datasets.mnist.load_data()
```

```python
print("train images")
print("amount", len(train_images))

print(len(train_images[0]))
print(len(train_images[0][0]))
print(train_images)
```

```python
print("train labels")
print(train_labels)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax),
])

model.summary()
```

```python
model.compile(
    optimizer="adam",
    loss=keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"],
)
```

```python
model.fit(train_images, train_labels, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("loss", test_loss)
print("acc", test_acc)
```

```python
img = image.load_img("number.png", color_mode="grayscale", target_size=(28, 28))
```

```python
# Convert the image to a numpy array
img_array = image.img_to_array(img)

print(img_array, len(img_array), len(img_array[0]))

# Normalize the image array
img_array = img_array / 255.0  # Model dependent: some models expect inputs in range [0, 1], others in [-1, 1]

print(img_array, len(img_array), len(img_array[0]))

# Add a batch dimension
img_array = np.expand_dims(img_array, axis=0)  # Model expects shape (batch_size, height, width, channels)

img_array
```

```python
predictions = model.predict(img_array)
print(predictions)
```

```python
# For classification, you might want to get the class with the highest probability
predicted_class = np.argmax(predictions, axis=-1)
print("Predicted Class:", predicted_class)
```
